### Scatter

```python
import numpy as np
```

```python
x = np.arange(1,11)
y=x.copy()
```

```python
plt.figure() 
plt.scatter(x,y)
```

```python
plt.figure()
colors = ['green']* (len(x)-1)
colors.append('red')
plt.scatter(x,y,s=200, c=colors)
```

### Line Plot 

```python
plt.figure()
plt.plot(x,'-o', x**2,'-o')
b = [50,53,55,54]
plt.plot(b, '--r')
```

```python
a = plt.gca()
```

```python
a.fill_between(range(len(x)),
              x,
              x**2,
              color = 'blue',
              alpha = 0.25)
```
