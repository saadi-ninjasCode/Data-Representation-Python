```python
import numpy as np
```

```python
data = np.arange(1,11)
```

```python
fig, ax = plt.subplots(3,3, sharex=True , sharey=True)
```

```python
ax[1,2].plot(data)
```

```python
#label the all figure ticks
for label in plt.gcf().get_axes():
    for l in label.get_xticklabels() + label.get_yticklabels():
        l.set_visible(True)
```

```python
plt.gcf().canvas.draw()
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, sharex=True)
ax = [ax1,ax2,ax3,ax4]
for a in range(len(ax)):
    sample = 10** (a+1)
    rand_sample = np.random.normal(0.0, 1, size=sample)
    ax[a].hist(rand_sample, bins=1000)
    
```

```python
from matplotlib.gridspec import GridSpec
```

```python
plt.figure()
gs = GridSpec(3,3)
top = plt.subplot(gs[0,:])
side = plt.subplot(gs[1,:-1])
bottom = plt.subplot(gs[-1,:-1])
last = plt.subplot(gs[1:, 2:])
```

```python
X = np.random.normal(0.0, 1.0, 10000)
Y = np.random.random(10000)
```

```python
top.hist2d(X,Y)
```

```python
side.hist(X)
bottom.hist(Y)
```
